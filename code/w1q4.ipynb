```python
import numpy as np
import time

n: int = 1000000
d: int = 10

# Generate data and query
X: np.ndarray = np.random.rand(n,d)
q: np.ndarray = np.random.rand(d)

# Vectorised Approach
start: float = time.perf_counter()
vectorised_d: np.ndarray = np.sqrt(np.sum(X**2, axis=1) - 2 * X.dot(q) + np.sum(q**2))
end: float = time.perf_counter()

print(f"Vectorised Approach: {(end-start):.4f} seconds")

# Naive Approach
start: float = time.perf_counter()
naive_d: np.ndarray = np.array([np.linalg.norm(row - q) for row in X])
end: float = time.perf_counter()

print(f"Naive Approach: {(end-start):.4f} seconds")
```
